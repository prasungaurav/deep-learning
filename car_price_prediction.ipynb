```python
df = pd.read_csv("CarPrice_Assignment.csv")
```

```python
df.isnull().sum()
```

```python
df.drop(["car_ID"], axis=1, inplace=True)


```

```python
df.head(30)
```

```python
labels = df["CarName"].nunique()
```

```python
print(labels)
```

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
```

```python
label_encoder = LabelEncoder()
```

```python
categorical_cols = ["CarName", "fueltype","aspiration","doornumber","carbody","drivewheel","enginelocation","fuelsystem","enginetype","cylindernumber"]
```

```python
for col in categorical_cols:
 df[col] = label_encoder.fit_transform(df[col])
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.iloc[:, :24], df.iloc[:, 24],test_size=0.2)
```

```python
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

y_scaler = MinMaxScaler()
y_train = y_scaler.fit_transform(y_train.values.reshape(-1, 1))
y_test = y_scaler.transform(y_test.values.reshape(-1, 1))
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense
```

```python
model=Sequential([
     Dense(64,activation='relu',input_shape=(24,)),
     Dense(32,activation='relu'),
     Dense(16,activation='relu'),
     Dense(1,activation='linear')
])
```

```python
model.compile(optimizer='adam',loss='mse',metrics=['mse'])
```

```python
model.fit(x_train,y_train,epochs=100,batch_size=64,validation_data=(x_test,y_test))
```

```python
model.evaluate(x_train,y_train,batch_size=64)
```

```python
import joblib
```

```python
joblib.dump(model, "car_price.pkl")
```
